# Text Preprocessing

## Tokenizer with Keras

```python
import numpy as np
import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
sentence = ['Hello! My name is Shanin Hossain', 'I read in B.SC. in CSE', 'I love Programming', 
           'My favourite programming language is Python', 'Now i am working on Deep Learning']

model = Tokenizer(num_words = 100, oov_token = '<OOV>')    # num_words/Vocabulary size | OOV(out of vocabulay) 
```

```python
model.fit_on_texts(sentence)
words = model.word_index    # all word numerical value in dictionary
print(words)    # all word in lower_Case
```

```python
sequence = model.texts_to_sequences(sentence)
print(sequence)
```

```python
test_text = ['My Youtube channel name is CS Inside']
sequence1 = model.texts_to_sequences(test_text)
print(sequence1)
#  2=My, 7=name, 3=is but index value of Youtube, channel, CS, Inside are not writtern in the seq because these words are not present in main sentence. If we add we need OOV function
# replace new word index with value 1
```

```python
# make all the sentence in the same length we need pad_sequence
padding = pad_sequences(sequence, padding = 'post', maxlen = 8, truncating = 'post')
print(padding)
# post = putting all the zeroes at the end
#  maxlen = maximum length of all the list will be given number(not important)   
# truncating = if a sentence is bigger than length, then delete at the end
```
