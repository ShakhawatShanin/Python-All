# Text Feature Extraction

## Countvectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
text0 = ["Hello! My name is Shanin", "I read in CSE", "I love play Cricket"]
text1 = ["My favourite programming language is Python which is powerfull"]

# create the transform
vectorizer = CountVectorizer()

# tokenize and build vocab
vecFit = vectorizer.fit(text0)

print(vectorizer.vocabulary_)
print(vectorizer.get_feature_names())     # all unique word/tokenization

# Index   Word
#   0     hello
#   1     hossain
#   2     is
#   3     my
#   4     name
#   5     shanin
```

```python
vecFit = vectorizer.transform(text0)
print(vecFit.shape)   # Total sentence & unique word number
print(vecFit)    # position of non-zero
print(vecFit.toarray())     
# OR
df = pd.DataFrame(vecFit.toarray(), columns = vectorizer.get_feature_names())
print(df)
```

```python
# encode document
newVector = vectorizer.transform(text1)

# summarize encoded vector
print(newVector.toarray())    # common words are 1 and rest of the element become 0
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
text2 = ["Sky is blue", "Sun is bright"]
```

```python
tfidf_vec = TfidfVectorizer()

tfidfFit = tfidf_vec.fit(text2)

print(tfidf_vec.vocabulary_)
print(tfidf_vec.get_feature_names()) 
```

```python
tfidfFit = tfidf_vec.transform(text2)
print(tfidfFit.shape)   # Total sentence & unique word number
print(tfidfFit)    # position of non-zero
print(tfidfFit.toarray())     
# OR
df = pd.DataFrame(tfidfFit.toarray(), columns = tfidf_vec.get_feature_names())
print(df)
```
